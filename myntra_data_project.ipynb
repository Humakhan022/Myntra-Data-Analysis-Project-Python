# **<font color="$5831bc" face="Comic Sans MS">Bar Plot: Product Count of Top 10 Brands</font>**

```
top_10_models = df['brand_name'].value_counts().nlargest(10).index
df_top_10_models = df[df['brand_name'].isin(top_10_models)]

plt.figure(figsize = (12, 6))
ax = sns.countplot(x = 'brand_name',data = df_top_10_models, order=top_10_models)
for bars in ax.containers:
  ax.bar_label(bars)
plt.title("Count of Top 10 brand_name", fontsize = 15,color='Blue');
plt.xticks(rotation=45, ha='left')
plt.tight_layout()
plt.show()
```

### **<font color=darkred>OBSERVATION**
* **This bar chart visualizes the top 10 brands based on product count in the dataset.**
* **The chart clearly shows that a few brands have significantly higher product volumes, while others contribute comparatively fewer products.**
* **Such insights help in understanding brand dominance, inventory distribution, and market concentration, which are important for data-driven business decisions.**

# **<font color="$5831bc" face="Comic Sans MS">Bar Plot: Percentage Share of Top 10 Brands by Product Count</font>**

```
top_10_models = df['brand_name'].value_counts().nlargest(10).index
df_top_10_models = df[df['brand_name'].isin(top_10_models)]

plt.figure(figsize = (12, 6))
ax = sns.countplot(x = 'brand_name',data = df_top_10_models, order=top_10_models)

total = len(df_top_10_models)
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width() / 2.
    y = p.get_height()
    ax.annotate(percentage, (x, y), ha='center', va='bottom')

plt.title("Count of Top 10 brand_name with Percentage", fontsize = 15,color='Red');
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### **<font color=darkred>OBSERVATION**
* **This bar chart highlights the top 10 brands based on product count, along with their percentage contribution to the total products of these brands.**
* **A few brands contribute a higher percentage share, clearly indicating brand dominance, while others hold a relatively smaller percentage.**
* **Displaying percentages on the bars makes it easier to compare brand presence, understand market concentration, and identify leading brands at a glance.**

# **<font color="$5831bc" face="Comic Sans MS">Line Plot: How Product Price Relates to Average Number of Ratings</font>**

```
plt.figure(figsize=(15, 6))

df_line = df.groupby('price')['number_of_ratings'].mean().reset_index()
sns.lineplot(x='price', y='number_of_ratings', data=df_line, color='green')
plt.title('Line Plot: price vs number_of_ratings')
plt.xlabel('price')
plt.ylabel('number_of_ratings')
plt.show()
```

### **<font color=darkred>OBSERVATION**
* **I analyzed the relationship between product price and the average number of ratings using a line plot.**
* **The visualization shows how customer engagement (number of ratings) varies with pricing.**
* **Insights from this analysis can help brands understand pricing impact on popularity and customer response, supporting better pricing and marketing strategies.**

# **<font color="$5831bc" face="Comic Sans MS">Box Plot Analysis: Pricing Patterns of Top Brands</font>**

```
plt.figure(figsize=(15, 6))

# Calculate top 10 brands based on product count
top_10_brands = df['brand_name'].value_counts().nlargest(10).index.tolist()

# Filter data for top 10 brands
df_top_10 = df[df['brand_name'].isin(top_10_brands)]

# Box plot for MRP distribution of top 10 brands
sns.boxplot(
    x='brand_name',
    y='MRP',
    data=df_top_10,
    palette='Set2'
)

plt.title('Box Plot: MRP Distribution Across Top 10 Brands')
plt.xlabel('Brand Name')
plt.ylabel('MRP')

plt.xticks(rotation=45)
plt.show()
```

### **<font color=darkred>OBSERVATION**
* **I analyzed the MRP (Maximum Retail Price) distribution of the top 10 brands using a box plot to understand their pricing patterns.**
* **The visualization shows variation in pricing, with some brands having a wide price range while others maintain consistent pricing.**
* **Such insights are valuable for competitive pricing strategy, market positioning, and identifying premium vs. budget brands.**

# **<font color="$5831bc" face="Comic Sans MS">Bar Plot: Ratings of the Bottom 70 Brands by Customer Feedback</font>**

```
plt.figure(figsize = (20, 6))
plt.grid()
plt.xticks(rotation = 90)
plt.ticklabel_format(style = 'plain')
df.sort_values("ratings", axis = 0, ascending = True, inplace = True)
sns.barplot(x = df["brand_name"][:70], y = df["ratings"], palette = "twilight_shifted_r");
```

### **<font color=darkred>OBSERVATION**

# **<font color="$5831bc" face="Comic Sans MS">Bar Plot: Total Discount Percent Contribution by Top 10 Brands</font>**

```
top_10_brands = df.groupby('brand_name')['discount_percent'].sum().nlargest(10).sort_values(ascending=False).index
df_top_10 = df[df['brand_name'].isin(top_10_brands)]
#display(df_top_10)

plt.figure(figsize = (8, 6))
plt.ticklabel_format(style = 'plain')
sns.barplot(x = df_top_10["brand_name"], y = df_top_10["discount_percent"], palette = "tab20");
plt.title('Top 10 brand_name by Total discount_percent')
plt.xlabel('brand_name')
plt.ylabel('Total discount_percent')
plt.xticks(rotation=65, ha='right')
plt.tight_layout()
plt.show()
```

### **<font color=darkred>OBSERVATION**
* **As part of my data analysis on brand discounts, I identified the top 10 brands
contributing the highest total discount percentages.**
* **The bar chart clearly shows that a few brands dominate the discount landscape, highlighting their aggressive pricing strategies to attract more customers.**
* **Such insights are valuable for marketing strategy, competitive analysis, and understanding brand behavior in the market.**
